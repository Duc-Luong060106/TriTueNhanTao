# Máy hút bụi - MODEL-BASED-REFLEX-AGENT

percept: Ma trận mxn, các giá trị mô tả: 
- 0: ô không có bụi 
- 1: ô có bụi 
- 2: ô vật cản 
 
Kết quả: Chuỗi các bước di chuyển, gặp ô có bụi sẽ hút bụi. 

Dừng lại khi không còn bước di chuyển hợp lệ (tất cả các nước đi có thể đều quay lại trạng thái đã xuất hiện trước đó) hoặc bụi đã được hút hết. 

```python
import random

def is_cleaned(percept):
    for i in range(m):
        for j in range(n):
            if percept[i][j] == 1:
                return False

    return True

def print_matrix(percept, state):
    x, y = state
    for i in range(m):
        for j in range(n):
            if i == x and j == y:
                print(f"[{percept[i][j]}]", end="")
            else:
                print(f" {percept[i][j]} ", end="")
        
        print("")

def to_tuple(percept):
    return tuple(tuple(row) for row in percept)

def copy_percept(percept):
    return [row[:] for row in percept]

def model(percept, state, visited):
    x, y = state
    blocked = []

    if x > 0 and (to_tuple(percept), (x-1, y)) in visited:
        blocked.append("U")

    if x < m - 1 and (to_tuple(percept), (x+1, y)) in visited:
        blocked.append("D")

    if y > 0 and (to_tuple(percept), (x, y-1)) in visited:
        blocked.append("L")

    if y < n - 1 and (to_tuple(percept), (x, y+1)) in visited:
        blocked.append("R")

    return blocked

def update_state(percept, state, visited):
    blocked = model(percept, state, visited)
    
    return state, blocked       

def move_rule(percept, state, blocked):
    x, y = state
    moves = []

    if x > 0 and "U" not in blocked and percept[x-1][y] != 2:
        moves.append("U")
    if x < m - 1 and "D" not in blocked and percept[x+1][y] != 2:
        moves.append("D")
    if y > 0 and "L" not in blocked and percept[x][y-1] != 2:
        moves.append("L")
    if y < n - 1 and "R" not in blocked and percept[x][y+1] != 2:
        moves.append("R")

    if not moves:
        return "NOT FOUND"
    else:
        return random.choice(moves)

def rule_match(percept, state, blocked):
    x, y = state
    
    if percept[x][y] == 1:
        return (True, move_rule(percept, state, blocked))
    else:
        return (False, move_rule(percept, state, blocked))

def action(result_rule, percept, state):
    new_percept = copy_percept(percept)
    x, y = state
    clean, move = result_rule

    if clean:
        new_percept[x][y] = 0

    if move == "U":
        x -= 1

    elif move == "D":
        x += 1

    elif move == "L":
        y -= 1

    elif move == "R":
        y += 1

    return new_percept, (x, y)

def model_based_reflex_agent(percept, state, visited):
    state, blocked = update_state(percept, state, visited)

    result_rule = rule_match(percept, state, blocked)

    if result_rule[1] == "NOT FOUND":
        return percept, state, result_rule

    new_percept, new_state = action(result_rule, percept, state)

    return new_percept, new_state, result_rule

def solve(percept, state):
    visited = set()
    step = 0

    while True:
        print(f"\nBước: {step}")
        print_matrix(percept, state)

        if is_cleaned(percept):
            print("Phòng đã được dọn sạch")
            break

        visited.add((to_tuple(percept), state))

        percept, state, actions = model_based_reflex_agent(percept, state, visited)

        if actions[1] == "NOT FOUND" or step == 1000:
            if action[0]:
                print("Dọn dẹp ô hiện tại, Di chuyển: ", end="")
            print("Không tìm được đường đi tiếp theo")
            break

        if actions[0]:
            print(f"Dọn dẹp ô hiện tại và di chuyển: {actions[1]}")
        else:
            print(f"Không dọn dẹp ô hiện tại và di chuyển: {actions[1]}")
        step += 1

percept = [
    [1,0,1],
    [0,2,0],
    [1,0,1],
    [2,1,0]
]

m = len(percept)
n = len(percept[0])

state = 0, 0
solve(percept, state)
```
