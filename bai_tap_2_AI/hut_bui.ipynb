    Các ô đánh số: 1 khi có bụi và 0 khi không có bụi.

S: Cơ chế nhận biết môi trường truyền vào (ma trận). Nếu ngoài thực tế thì có thêm camera, cơ chế nhận biết được bụi, vật cản để nhận biết.

```python
import random

def  interpret_input(percept):
    global x, y
    return x, y

def clean_action(x, y):
    percept[x][y] = 0
    print(f"Robot đã hút bụi ô ({x}, {y})")

def move_action(x, y):
    moves = []
    old_x, old_y = x, y
    if x > 0:
        moves.append("up")
    if x < m - 1:
        moves.append("down")
    if y > 0:
        moves.append("left")
    if y < n - 1:
        moves.append("right")

    move = random.choice(moves)
    match move:
        case "up":
            x -= 1
        case "down":
            x += 1
        case "left":
            y -= 1
        case "right":
            y += 1
    print(f"Robot di chuyển từ ({old_x}, {old_y}) sang ô ({x}, {y})")
    return x, y

def rule_match(state):
    x, y = state

    if percept[x][y] == 1:
        clean_action(x, y)
        return move_action(x, y)
    else:
      return move_action(x, y)
    
def simple_reflex_agent(percept):
    state = interpret_input(percept)

    x, y = rule_match(state)

    return x, y

def solve():
    global x, y 

    visited = set()
    print(f"Vị trí ban đầu: ({x}, {y})")

    while True:
        if (x, y) in visited:
            print("Thuật toán dừng!")
            break

        visited.add((x, y))

        print(percept)

        x, y = simple_reflex_agent(percept)

percept = [
    [1,0,1],
    [0,1,0],
    [1,0,1]
]

m = len(percept)
n = len(percept[0])

x, y = 0, 0
solve()
```
