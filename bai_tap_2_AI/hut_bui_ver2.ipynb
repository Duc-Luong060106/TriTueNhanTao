  
Quy ước percept (ma trận MxN) mô tả sơ đồ bố trí phòng:  
  - 0: ô không có bụi  
  - 1: ô có bụi  
  - 2: ô bị cản không thể di chuyển (nội thất)  
  

```python
import random

class Percept:
    def __init__(self, matrix, x, y):
        self.matrix = matrix
        self.x = x
        self.y = y
    
    def print_matrix(self):
        for i in range(len(matrix)):
            for j in range(len(matrix[0])):
                if self.x == i and self.y == j:
                    print('#', end=' ')
                else:
                    print(self.matrix[i][j], end=' ')
            print('')
    
    def is_cleaned(self):
        for i in range(len(matrix)):
            for j in range(len(matrix[0])):
                if self.matrix[i][j] == 1:
                    return False
        
        return True
        

def  interpret_input(percept: Percept):
    return percept.x, percept.y

def clean_action(percept: Percept, x, y):
    percept.matrix[x][y] = 0
    print(f"Robot đã hút bụi ô ({x}, {y})")

def move_action(percept: Percept, x, y):
    moves = []

    if x > 0 and percept.matrix[x-1][y] != 2:
        moves.append("up")
    if x < m - 1 and percept.matrix[x+1][y] != 2:
        moves.append("down")
    if y > 0 and percept.matrix[x][y-1] != 2:
        moves.append("left")
    if y < n - 1 and percept.matrix[x][y+1] != 2:
        moves.append("right")

    move = random.choice(moves)
    match move:
        case "up":
            x -= 1
        case "down":
            x += 1
        case "left":
            y -= 1
        case "right":
            y += 1
    print(f"Robot di chuyển từ ({percept.x}, {percept.y}) sang ô ({x}, {y})")
    return x, y

def rule_match(state, percept: Percept):
    x, y = state

    if percept.matrix[x][y] == 1:
        clean_action(percept ,x, y)
        return move_action(percept, x, y)
    else:
      return move_action(percept, x, y)
    
def simple_reflex_agent(percept):
    state = interpret_input(percept)

    x, y = rule_match(state, percept)

    return x, y

def solve():

    visited = set()
    print(f"Vị trí ban đầu: ({percept.x}, {percept.y})")

    while True:
        if (percept.x, percept.y) in visited or percept.is_cleaned():
            print("Thuật toán dừng!")
            break

        visited.add((percept.x, percept.y))

        percept.print_matrix()

        percept.x, percept.y = simple_reflex_agent(percept)

        print('')

matrix = [
    [1,0,2,1],
    [0,2,1,0],
    [1,0,0,1],
    [2,1,0,0]
]
m = len(matrix)
n = len(matrix[0])

percept = Percept(matrix, 0, 0)

solve()
```
