```python
TARGET = [
    [1, 2, 3], 
    [4, 5, 6], 
    [7, 8, 0]
]

def interpret_input(percept):
    for i in range(3):
        for j in range(3):
            if percept[i][j] == 0:
                return (i, j)

def move(percept, state, action):
    new_percept = percept.copy()
    i, j = state

    if action == "UP":
        new_percept[i][j], new_percept[i-1][j] = new_percept[i-1][j], new_percept[i][j]

    elif action == "DOWN":
        new_percept[i][j], new_percept[i+1][j] = new_percept[i+1][j], new_percept[i][j]

    elif action == "LEFT":
        new_percept[i][j], new_percept[i][j-1] = new_percept[i][j-1], new_percept[i][j]

    elif action == "RIGHT":
        new_percept[i][j], new_percept[i][j+1] = new_percept[i][j+1], new_percept[i][j]

    return new_percept

def rule_match(state):
    i, j = state
    i_goal, j_goal = 2, 2

    if i < i_goal:
        return "DOWN"
    if i > i_goal:
        return "UP"
    if j < j_goal:
        return "RIGHT"
    if j > j_goal:
        return "LEFT"
    
    return "NOT FOUND"

def simple_reflex_agent(percept):
    state = interpret_input(percept)

    action = rule_match(state)
    if action == "NOT FOUND":
        return ""
    percept = move(percept, state, action)

    return action


test_state = [
    [1, 2, 3], 
    [4, 0, 6], 
    [7, 5, 8] 
]

while True:
    res = simple_reflex_agent(test_state)

    if res == "":
        break
    else:
        print(res)
        print(test_state)

```
