## Giải bài toán 8-puzzles bằng thuật toán Breadth First Search (BFS) 
- Nhận vào: trạng thái ban đầu (start) và trạng thái mục tiêu (goal) 
- Kết quả: tìm ra chuỗi các bước di chuyển để từ start thành goal hoặc Failure (Nếu không tìm thấy)

```python
from collections import deque

class Node:
    def __init__(self, state, parent, action, path_cost):
        self.state = state
        self.parent = parent
        self.action = action
        self.path_cost = path_cost

def get_path(node):
    path = []
    while node.parent != "":
        path.append(node.action)
        node = node.parent
    return list(reversed(path))

def gen_actions(state):
    found = False
    for x in range(3):
        for y in range(3):
            if state[x][y] == 0:
                found = True
                break
        if found:
            break
            
    
    actions = []
    if x > 0:
        up_state = [row[:] for row in state]
        up_state[x][y], up_state[x-1][y] = up_state[x-1][y], up_state[x][y]

        actions.append(("U", up_state))

    if x < 2:
        down_state = [row[:] for row in state]
        down_state[x][y], down_state[x+1][y] = down_state[x+1][y], down_state[x][y]

        actions.append(("D", down_state))

    if y > 0:
        left_state = [row[:] for row in state]
        left_state[x][y], left_state[x][y-1] = left_state[x][y-1], left_state[x][y]

        actions.append(("L", left_state))
    
    if y < 2:
        right_state = [row[:] for row in state]
        right_state[x][y], right_state[x][y+1] = right_state[x][y+1], right_state[x][y]

        actions.append(("R", right_state))
        
    return actions


def breadth_first_search(initial, goal_test):
    node = Node(initial, "", "", 0)
    if goal_test == node.state:
        return get_path(node)
    
    frontier = deque()
    frontier.append(node)
    reached = set()

    while frontier:
        node = frontier.popleft()
        reached.add(tuple(tuple(row) for row in node.state))

        if goal_test == node.state:
            return get_path(node)

        for action, child_state in gen_actions(node.state):
            child = Node(child_state, node, action, node.path_cost + 1)

            if child.state not in [n.state for n in frontier] and tuple(tuple(row) for row in child.state) not in reached:
                frontier.append(child)

    return "Failure"

start = [[1, 2, 3],
         [8, 0, 4],
         [7, 6, 5]]

goal = [[2, 8, 1],
        [0, 4, 3],
        [7, 6, 5]]

print("Start:")
for row in start:
    print(row)

print("\nGoal:")
for row in start:
    print(row)

print("Chuỗi bước di chuyển tìm được: ", breadth_first_search(start, goal))
        
```
